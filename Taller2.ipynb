```python
#variable para capturar la venta 
venta1 = float(input("¿Cuanto fue el total de la venta1? $"))
venta2 = float(input("¿Cuanto fue el total de la venta2? $"))
venta3 = float(input("¿Cuanto fue el total de la venta3? $"))

#Calculo de la comisión 
comision1 = venta1 * 0.10
comision2 = venta2 * 0.10
comision3 = venta3 * 0.10
  
#calcular total comisiones 
comisiont=comision1+comision2+comision3


sueldo = float(input(" Ingrese su salario $ "))

total = sueldo + comisiont
print("Su comision es = ${:,.0f} y el total de su salario mas comision es= {:,.0f}".format(comisiont, total))
```

```python
#variable para capturar la compra 
compra = float(input("¿Cuanto fue el total de la compra? $"))

#calcular descuento 
descuento=compra * 0.15

#calcular total
total= compra - descuento

print("El total de su compra es ${:,.0f}".format(total))

```

```python
# variables para capturar notas 
nota1 = float(input("¿Cuanto fue la nota del primer parcial? $"))
nota2 = float(input("¿Cuanto fue la nota del segundo parcial? $"))
nota3 = float(input("¿Cuanto fue la nota del tercer parcial? $"))

#calcular valor final 
final = (nota1+nota2+nota3)/3


print("Su nota final es ${:,.0f}".format(final))
```

```python
# variables para total de estudiantes mujeres
mujeres= float(input("Ingrese la cantidad de mujeres que hay en su grupo: "))
# variables para total de estudiantes hombres
hombres= float(input("Ingrese la cantidad de hombres que hay en su grupo: "))

total= mujeres+hombres 

#calculo porcentaje 
est_muj= ((mujeres*100)/total)
est_hom= ((hombres*100)/total)


print("El porcentaje de estudiantes mujeres es = {:,.0f}% y el porcentaje de estudiantes hombres es= {:,.0f}%".format(est_muj, est_hom))



```
