```python
#bucle para pares hasta 100

for p in range(0,102,2):
  print(p)
```

```python
# bucle tabla de multiplicar
numero = int(input("ingrese numero: "))
for m in range(1,11,1):
  print("{:,.0f} * {:,.0f} = {:,.0f} ".format(numero,m,(m*numero)))
```

```python
# bucle edad usuario
e = int(input("ingrese su edad: "))
for n in range(1,e+1,1):
  print(f"usted ha cumplido:= {n} años")
```

```python
# bucle para impares
n = int(input("Ingrese un numero: "))
for n in range(1,n+1,2):
  print(n,end=", ")
```

```python
# bucle para suma pares
acu = 0
for n in range(0,100+1,2):
  acu = acu + n
print(f"suma de numeros pares 0-100= {acu}")
```
