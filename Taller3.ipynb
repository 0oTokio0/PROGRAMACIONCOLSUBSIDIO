```python
#Capturar información del vendedor 
nombre= input("Ingrese su nombre : ")
edad= int(input("Ingrese su edad : "))
e_civil= input(" Ingrese estado civil: ")
hijos= int(input("Cuantos hijos tiene?: "))
basico= float(input("Ingrese su salario basico: "))
laborado=float(input("Ingrese la cantidad de días laborados: "))



#Calculo bono prepensión 
if (edad>55):
  prepension= basico * 0.05

  print("Tiene el beneficio de bono prepensión por valor: ${:,.0f}".format(prepension)) 
else:
    print(" No aplica bono  prepensión")



#Calculo paseo
if e_civil=="Casado" and hijos>1:
    print("¡Es beneficiario del paseo familiar!")
  
else:
      print("No aplica paseo familiar")

#Calculo comisiones
if basico==1000000 or basico<=1500000:
    comision=basico * 0.02
    print("Su comisión es : ${:,.0f}".format(comision)) 
elif basico==1500001 or basico<=2000000:
    comision=basico * 0.05
    print("Su comisión es : ${:,.0f}".format(comision)) 
else:
      print("No aplica comision") 


#Calculo bono alimentacion
if laborado>20 and basico<1000000:
    print("aplica bono de alimentacion")
else:
    print("No aplica bono de alimentacion")




```
